```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Sample data: Students' Math and Science scores
data = {
    'Math': [75, 80, 82, 60, 58, 55, 30, 28, 35, 90, 92, 88],
    'Science': [78, 85, 84, 55, 53, 52, 25, 20, 30, 91, 93, 89]
}

# Create DataFrame
df = pd.DataFrame(data)

# Prepare the data for clustering
X = df.values

# Plot dendrogram to determine cluster count
linked = linkage(X, method='ward')

plt.figure(figsize=(10, 6))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title("Dendrogram - Student Clustering")
plt.xlabel("Students")
plt.ylabel("Euclidean Distance")
plt.show()

# Apply Agglomerative Clustering (let's assume 3 clusters)
model = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
df['Cluster'] = model.fit_predict(X)

# Print cluster assignments
print(df)

# Plot the clustered data
plt.figure(figsize=(8, 6))
plt.scatter(df['Math'], df['Science'], c=df['Cluster'], cmap='Set1')
plt.title("Hierarchical Clustering of Students")
plt.xlabel("Math Score")
plt.ylabel("Science Score")
plt.grid(True)
plt.show()
```
